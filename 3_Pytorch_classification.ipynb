```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./data/credit-a.csv', header = None)
```

```python
data.head()
```

The first 15 columns are features, the last one is target

```python
X = data.iloc[:, :-1]
X.shape
```

```python
Y = data.iloc[:, -1]
Y.shape
```

```python
Y.value_counts()
```

```python
#coonvert Y to 0 and 1
Y.replace(-1,0, inplace = True)
```

```python
Y.value_counts()
```

```python
X = torch.from_numpy(X.values).type(torch.FloatTensor)
Y = torch.from_numpy(Y.values.reshape(-1, 1)).type(torch.FloatTensor)
```

```python
X.shape
```

```python
Y.shape
```

Add a layer of sigmoid for regression, it will become classification.

```python
from torch import nn
model = nn.Sequential(
    nn.Linear(15, 1024),
    nn.Linear(1024,1),
    nn.Sigmoid()
)
```

```python
model
```

```python
#loss function is binary cross entropy
loss_fn = nn.BCELoss()
```

```python
opt = torch.optim.SGD(model.parameters(), lr = 0.001)
```

```python
batch_size = 32
steps = 653//32
```

```python
for epoch in range(1000):
    for batch in range(steps):
        start = batch*batch_size
        end = start + batch_size
        x = X[start:end]
        y = Y[start:end]
        y_pred = model(x)
        print(f'In bactch {batch}, we have of {x.shape[0]} features  and {y.shape[0]} targets.')
        loss = loss_fn(y_pred, y)
        opt.zero_grad()
        loss.backward()
        opt.step()
```

```python
model.state_dict()
```

```python
#calculate accuracy
((model(X).data.numpy()>0.5) == Y.numpy()).mean()
```
